### Quiz 4

```python
G = nx.Graph()
```

```python
G.add_edges_from([
    ('A', 'B'), 
    ('A', 'C'),
    ('A', 'F'),
    ('B', 'C'),
    ('B', 'D'),
    ('B', 'E'),
    ('F', 'C'),
    ('F', 'E'),
    ('C', 'E'),
    ('E', 'D')
])
```

```python
plt.figure(figsize=(10, 10))
nx.draw_networkx(G)
```

```python
degrees = G.degree()
degree_values = sorted(set(degrees.values()))
histogram = [list(degrees.values()).count(i)/float(nx.number_of_nodes(G)) for i in degree_values]
```

```python
G = nx.Graph()
```

```python
G.add_edges_from([
    ('C', 'G'), 
    ('C', 'A'),
    ('G', 'D'),
    ('A', 'D'),
    ('A', 'E'),
    ('D', 'B'),
    ('D', 'H'),
    ('D', 'E'),
    ('E', 'H'),
    ('H', 'F')
])
```

```python
plt.figure(figsize=(10, 10))
nx.draw_networkx(G)
```

```python
common_neigh = [(e[0], e[1], len(list(nx.common_neighbors(G, e[0], e[1])))) for e in nx.non_edges(G)]
```

```python
common_neigh = [x for x in common_neigh if any(y == 'H' for y in x[:2])]
```
